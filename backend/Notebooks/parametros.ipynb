## Lectura de datos Parametricos

```python
import pandas as pd

# Ruta de los archivos
file_path_parameters = '../dataset/datosParametros.xlsx'

# Leer el archivo de datos de parámetros
df_parameters = pd.read_excel(file_path_parameters)

# Verificar las primeras filas del DataFrame de parámetros
df_parameters.head()
```

```python
# Renombrar columnas del DataFrame de parámetros
df_parameters.columns = [
    'Parameter',
    'ParameterType',
    'Unit',
    'Cat3_D1',
    'Cat3_D2',
    'QChoc1',
    'RHuat6',
    'RSalc1',
    'RUrub0',
    'RUrub1',
    'RVilc1',
    'RVilc10',
    'RVilc11',
    'RVilc13',
    'RVilc14',
    'RVilc18',
    'RVilc20',
    'RVilc21',
    'RVilc24',
    'RVilc26',
    'RVilc27',
    'RVilc3',
    'RVilc31',
    'RVilc4',
    'RVilc6',
    'RVilc8',
    'RYana3'
]
df_parameters.head()
```

```python
# Reemplazar valores no numéricos
df_parameters.replace('----', np.nan, inplace=True)
df_parameters.head()
```

```python
max_temp = df_parameters[['Cat3_D1', 'Cat3_D2']].replace({'<': '', '<=': '', '±': ''}, regex=True).apply(pd.to_numeric, errors='coerce').max().max()
min_temp = df_parameters[['Cat3_D1', 'Cat3_D2']].replace({'<': '', '<=': '', '±': ''}, regex=True).apply(pd.to_numeric, errors='coerce').min().min()

# Función para extraer los límites de un valor
def extract_limits(value):
    if isinstance(value, str):
        if '<=' in value:
            return 0, float(value.replace('<=', '').replace(',', '.')) - 0.1
        elif '<' in value:
            return 0, float(value.replace('<', '').replace(',', '.'))
        elif '±' in value:
            value_num = float(value.replace('±', '').replace(',', '.'))
            return max_temp - value_num, max_temp + value_num
        else:
            return np.nan, np.nan
    return np.nan, np.nan

# Aplicar la extracción de límites para las columnas Cat3_D1 y Cat3_D2
df_parameters['Cat3_D1_min'], df_parameters['Cat3_D1_max'] = zip(*df_parameters['Cat3_D1'].apply(extract_limits))
df_parameters['Cat3_D2_min'], df_parameters['Cat3_D2_max'] = zip(*df_parameters['Cat3_D2'].apply(extract_limits))

# Mostrar las primeras filas del DataFrame con los nuevos límites
df_parameters[['Cat3_D1', 'Cat3_D1_min', 'Cat3_D1_max', 'Cat3_D2', 'Cat3_D2_min', 'Cat3_D2_max']].head()
```

```python
import pandas as pd
import numpy as np

# Función para limpiar y convertir los valores de las estaciones en formato numérico
def clean_station_values(value):
    if isinstance(value, str):
        if '<=' in value:
            return float(value.replace('<=', '').replace(',', '.')) - 0.1  # Reemplazar <= y restar 0.1
        elif '<' in value:
            return float(value.replace('<', '').replace(',', '.'))  # Reemplazar < y tomar el valor
        return float(value.replace(',', '.'))  # Reemplazar ',' por '.' en valores normales
    return np.nan  # Si no es una cadena, devuelve NaN

# Aplicar la limpieza a las columnas de las estaciones
station_columns = df_parameters.columns[5:-4]  
for col in station_columns:
    df_parameters[col] = df_parameters[col].apply(clean_station_values)

# Mostrar las primeras filas para verificar los cambios
df_parameters.head()
```
